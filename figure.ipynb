```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv('checkpoints/metrics_qlearning.csv')
df.columns
```

```python
fig, axs = plt.subplots(4, 1, figsize=(12, 9), sharex=True)

axs[0].plot(df['Episode'], df['MeanReward'])
axs[0].set_title('Mean Reward')

axs[1].plot(df['Episode'], df['MeanLoss'])
axs[1].set_title('Mean Loss')

axs[2].plot(df['Episode'], df['MeanQValue'])
axs[2].set_title('Mean Q Value')

axs[3].plot(df['Episode'], df['MeanLength'])
axs[3].set_title('Mean Length')

# Remove x-ticks from all subplots except the last one
for ax in axs[:-1]:
    plt.setp(ax.get_xticklabels(), visible=False)

# Add title
plt.suptitle('Metrics for Double Deep Q-Learning')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('checkpoints/metrics_actorcritic.csv')
df.columns
```

```python
fig, axs = plt.subplots(4, 1, figsize=(12, 9), sharex=True)

n = 50

axs[0].plot(df['Episode'][:n], df['MeanReward'][:n])
axs[0].set_title('Mean Reward')

axs[1].plot(df['Episode'][:n], df['MeanEntropyLoss'][:n], label='Entropy Loss', linestyle='--')
axs[1].plot(df['Episode'][:n], df['MeanActorLoss'][:n], label='Actor Loss', linestyle='--')
axs[1].plot(df['Episode'][:n], df['MeanCriticLoss'][:n], label='Critic Loss', linestyle='--')
axs[1].plot(df['Episode'][:n], df['MeanLoss'][:n], label='Total Loss')
axs[1].legend()
axs[1].set_title('Mean Loss')

axs[2].plot(df['Episode'][:n], df['MeanQValue'][:n])
axs[2].set_title('Mean Q Value')

axs[3].plot(df['Episode'][:n], df['MeanLength'][:n])
axs[3].set_title('Mean Length')

# Remove x-ticks from all subplots except the last one
for ax in axs[:-1]:
    plt.setp(ax.get_xticklabels(), visible=False)

# Add title
plt.suptitle('Metrics for Advantage Actor-Critic')

plt.tight_layout()
plt.show()
```
